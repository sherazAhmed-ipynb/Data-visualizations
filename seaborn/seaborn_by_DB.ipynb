## ***<span style="color:purple;">imports</span>***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
```

```python
crash_df.info()
```

## ***<center><span style="color:orange;">Distribution Plots</span></center>***

## ***Distribution Plot***

```python
sns.displot(crash_df['not_distracted'])
```

## ***joint plot***

```python
#kde = kernel density estimation
#kind = reg,kde,hex

sns.jointplot(x='speeding', y='alcohol',data=crash_df, kind='hex')
```

## ***<center><span style="color:red;">Pair Plots</span></center>***

```python
# sns.pairplot(crash_df)

tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df, hue='sex', palette='Greens')
```

## ***Rug Plot***

```python
sns.rugplot(tips_df['tip'])
```

## ***Styling***

```python
sns.set_style('whitegrid')

plt.figure(figsize=(8,4))

sns.set_context('paper',font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol',data=crash_df, kind='reg')
```

## ***<center><span style="color:navy;">Catgorical Plots</span></center>***

## ***Bar Plots***

```python
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

## ***Count Plot***

```python
sns.countplot(x='sex',data=tips_df)
```

## ***Box Plot*** 

```python
sns.boxenplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

## ***Strip Plot***

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day',y='total_bill', data=tips_df, jitter= True, hue='sex',dodge=True)
```

## ***Swarm Plot***

```python
sns.violinplot(x='day', y='total_bill',data=tips_df)
sns.swarmplot(x='day', y='total_bill',data=tips_df, color= 'White')
```

## ***Palettes***

```python
plt.figure(figsize=(7,5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips_df, hue='sex',palette='plasma')
```

## ***<center><span style="color:lightgreen;">Matrix Plots</span></center>***

## ***Heatmaps***

```python
plt.figure(figsize=(7,5))
sns.set_context('paper',font_scale=1.4)

#select only float64 data types columns
numric_col = crash_df.select_dtypes(include=['float64'])
crash_mx = numric_col.corr()

sns.heatmap(crash_mx, annot=True, cmap='viridis')
```

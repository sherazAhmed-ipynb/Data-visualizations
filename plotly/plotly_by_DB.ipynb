## ***<span style="color:purple;">imports</span>***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import seaborn as sns
import cufflinks as cf
import plotly.graph_objects as go
import plotly.express as px

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
arr = np.random.randn(50,4)
df = pd.DataFrame(arr,columns=['A','B','C','D'])
df.head(3)
```

## ***<span style="color:orange;">Line Plots</span>***

```python
df_stocks = px.data.stocks()
px.line(df_stocks, x='date',y='GOOG', labels={'x': 'Data', 'y':'Price'})

px.line(df_stocks, x='date', y=['GOOG','AAPL'],
        labels={'x':'Data','y':'Price'}, title="Apple vs. Google"
        )
```

## ***<span style="color:navy;">Bar Charts</span>***

```python
#get country column
df_us = px.data.gapminder().query("country == 'United States' ")
px.bar(df_us, x='year', y='pop')
```

```python
df_tips = px.data.tips()
px.bar(df_tips, x='day', y='tip', color='sex',
       title='Tips by sex on each day',
       labels={'tip': 'Tip Amount','day': 'Day of the week'}
       )
```

## ***<span style="color:navy;">Scatter Plots</span>***

```python
df_iris = px.data.iris()
px.scatter(df_iris, x='sepal_width', y='sepal_length', color='species',
           size='petal_length', hover_data=['petal_width']
           )
```

## ***<span style="color:navy;">Pie Charts</span>***

```python
df_asia = px.data.gapminder().query("year == 2007").query("continent == 'Aisa' ")
px.pie(df_asia, values='pop', names='country',
       title='Population of asian continent',
       color_discrete_sequence= px.colors.sequential.RdBu
       )
```

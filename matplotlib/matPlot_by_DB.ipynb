```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

%reload_ext autoreload
%autoreload 2
```

## ***<span style= "color:purple;">Functional plot</span>***

```python
xpts = np.linspace(0,5,10)
ypts = xpts**2
plt.plot(xpts,ypts)
plt.title('Days squared chart')
plt.xlabel('Days')
plt.ylabel('Days squared')

plt.show()
```

```python
fig_1 = plt.figure(figsize=(5,4), dpi=100)
# [left, bottom, width, hight]
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title('Days squared chart')
axes_1.plot(xpts,ypts, label = 'x/x^2')
axes_1.plot(ypts,xpts, label = 'x^2/x')
axes_1.legend(loc=1)
```

```python
ic_df = pd.read_csv('icecreamsales.csv')
ic_df = ic_df.sort_values(by='Temperature')
ic_df

np_arr = ic_df.values
np_arr

xpts = np_arr[:,0]
ypts = np_arr[:,1]
fig_2 = plt.figure(figsize=(6,4))
axes_2 = fig_2.add_axes([0,0,1,1])

axes_2.set_xlabel('Temerature')
axes_2.set_ylabel('icecream sales')
axes_2.set_title('Ice cream sales vs temperature')
axes_2.plot(xpts,ypts)
plt.show()
```
